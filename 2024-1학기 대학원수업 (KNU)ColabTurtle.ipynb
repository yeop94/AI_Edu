우리가 작성한 코드에서 `forward(size)` 와 `right(90)` 이 4번 반복되는 것을 알 수 있죠.

## 무한 반복

## while 반복문

`while` 뒤의 조건이 `False`(거짓)이 될 때까지 반복합니다.

우리는 `while`뒤에 조건으로 `True`(참)를 넣었고 조건이 `False`(거짓)로 변할리 없으므로 무한 반복하게 됩니다.

```python
size = 200
color = 'red'

initializeTurtle()
pencolor(color)

while True:
  for i in range(4):
    forward(size)
    right(90)
```

## 3번만 반복해서 그려보자

새로운 변수(counter)를 만들어서 `while`문을 몇 번 반복했는지 숫자로 저장해두죠.

그리고 몇 번 반복했는지 궁금하니까 `counter`를 출력하겠습니다.

### while문을 사용해서 3번 반복하기

````python
size = 200
color = 'red'
counter = 0

initializeTurtle()
pencolor(color)

while counter < 3:
  for i in range(4):
    forward(size)
    right(90)

  counter = counter + 1
````

### for문을 사용해서 3번 반복하기

````python
size = 200
color = 'red'

initializeTurtle()
pencolor(color)

for j in range(3):
  for i in range(4):
    forward(size)
    right(90)
````

## 반복할때마다 색깔 변경하기

우리가 방금 작성한 코드를 실행하면 빨간색 사각형을 총 3회 그리는데요.

아래와 같이 색깔을 변경하고 싶어요.

- 1회: 빨간색
- 2회: 노란색
- 3회: 초록색

이럴 경우에는 다음과 같은 조건이 붙어야겠죠.

1. 첫 번째 사각형을 그릴 때는 빨간색 펜을 사용한다
2. 두 번째 사각형을 그릴 때는 노란색 펜을 사용한다
3. 세 번째 사각형을 그릴 때는 초록색 펜을 사용한다

그렇다면 몇 번째 사각형을 그리는지 알아야하는데 그것은 바로 변수 `j`에 들어있죠.

사각형을 반복해서 그릴 때마다 `j`의 값이 0, 1, 2로 변합니다.

```python
size = 200
color = 'red'

initializeTurtle()
pencolor(color)

for j in range(3):
  for i in range(4):
    forward(size)
    right(90)
```

따라서 위의 조건을 프로그래밍으로 바꾸면:

1. 만약 `j == 0` 이면, `color = 'red'`
2. 만약 `j == 1` 이면, `color = 'yellow'`
3. 만약 `j == 2` 이면, `color = 'green'`

## if 조건문

1. `if` 만약 ㅇㅇ라면
2. `elif` if가 아니라면
4. `else` 이도저도 아니라면

````python
initializeTurtle()

pencolor('red')

for i in range(4):
    forward(size)
    right(90)

pencolor('yellow')

for i in range(4):
    forward(size)
    right(90)

pencolor('green')

for i in range(4):
    forward(size)
    right(90)
````

````python
size = 200

initializeTurtle()

for j in range(3):
  if j == 0:
    pencolor('red')
  elif j == 1:
    pencolor('yellow')
  elif j == 2:
    pencolor('green')

  for i in range(4):
    forward(size)
    right(90)
````

## 리스트를 사용해서 크기가 다른 사각형 그리기

`size` 변수도 숫자가 저장된 리스트 형태로 만들어서 다른 크기가 다른 사각형 3개를 만들 수 있지 않을까요?

````python
sizes = [50, 100, 200]
colors = ['red', 'yellow', 'green']

initializeTurtle()

for j in range(3):
  pencolor(colors[j])

  for i in range(4):
    forward(sizes[j])
    right(90)
````

## 함수(Function) 만들기

우리는 위에서 가장 간단한 사각형을 `for` 반복문을 사용해서 만들었었죠.

```python
size = 200
color = 'red'

initializeTurtle()
pencolor(color)

for i in range(4):
  forward(size)
  right(90)
```

사각형을 그리는 함수를 만들 건데요. 이렇게 작성합니다!

````python
def draw_rectangle():
  size = 200
  color = 'red'

  pencolor(color)

  for i in range(4):
    forward(size)
    right(90)
````

## 함수의 실행

````python
initializeTurtle()

draw_rectangle()
````

## 함수 입력 만들기

`size`와 `color`를 입력으로 받아 크기와 색깔이 다른 사각형을 편하게 그릴 수 있는 함수를 만들어 보아요!

<img src="https://i.imgur.com/DAmizAH.png" width="500px">

````python
def draw_rectangle(size, color):
  pencolor(color)

  for i in range(4):
    forward(size)
    right(90)
````

## 함수에 입력을 넣어서 실행하기

````python
initializeTurtle()

draw_rectangle(200, 'blue')
````

## 함수를 사용해서 색깔과 크기가 다른 사각형 그리기

위에서 배운 리스트와 함수를 적절히 섞어서 색깔과 크기가 다른 사각형을 그릴 수 있답니다.

````python
sizes = [50, 100, 200]
colors = ['red', 'yellow', 'green']

initializeTurtle()

for j in range(3):
  draw_rectangle(sizes[j], colors[j])
````
